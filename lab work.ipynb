```python
import pandas as pd
import random as r
import numpy as np
```

```python
random.seed("АДВИ_2023")
```

```python
index = list(range(1, 1001))
```

```python
np.random.shuffle(index)
```

```python
index
```

```python
list1 = ["группа_1", "группа_2", "группа_3", "группа_4", "группа_5"]
groups = []
```

```python
for i in range(1000):
    groups.append(r.choices(list1)[0])
```

```python
groups
```

```python
uniform = [random.randint(50, 100) for _ in range(1000)]
```

```python
uniform
```

```python
gauss_1 = [random.gauss(0, 1) for _ in range(1000)]

gauss_2 = [random.gauss(0, 12) for _ in range(1000)]

gauss_3 = [random.gauss(50, 9) for _ in range(1000)]
```

```python
gauss_2
```

```python
d = {'groups': groups,
     'uniform': uniform,
     'gauss_1': gauss_1,
     'gauss_2': gauss_2,
     'gauss_3': gauss_3}
```

```python
df = pd.DataFrame(index=index, data=d)
```

```python
df
```

```python
df.loc[df.index % 121 == 0, 'gauss_1'] = None
```

```python
df.loc[df['gauss_2'] % 1 > 0.95, 'gauss_2'] = None
```

```python
df.head(40)
```

```python
df['gauss_2'] = df['gauss_2'].fillna(df['gauss_2'].mean())
```

```python
df = df.dropna(subset=['gauss_1'])
```

```python
df
```

```python
df.gauss_2.mean()
```

```python
np.std(df.gauss_2)
```

```python
df.query('groups == "группа_5"').index.min()
```

```python
np.percentile(df.uniform, 90)
```

```python
df.query('uniform >= 95.0').uniform.count()
```

```python
df.groupby('groups', as_index=False).agg({'gauss_3': 'median'}).sort_values('gauss_3', ascending=False)
```

```python
df.groupby('groups', as_index=False).agg({'gauss_2': 'min'}).sort_values('gauss_2', ascending=False)
```

```python
df['new_gauss'] = df.gauss_1 * df.gauss_2 / df.gauss_3
```
