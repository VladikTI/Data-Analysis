```python
import pandas as pd
import seaborn as sns
import seaborn as plt

%matplotlib inline
```

```python
def read_n_agg(file_path):
    return pd.read_csv(file_path, sep=';') \
            .groupby('company').agg({'income': 'mean'})
```

```python
df = read_n_agg('companies.csv')
```

```python
taxi = pd.read_csv('taxi_peru.csv', sep=';', parse_dates=['start_at', 'end_at', 'arrived_at'])
```

```python
round(taxi.groupby('source').agg({'source': 'count'}).max() / taxi.source.count() * 100, 2)
```

```python
orders_source = pd.DataFrame(taxi.source.value_counts()) \
                .reset_index() \
                .rename(columns={'index': 'source', 'source': 'count'})
```

```python
orders_source
```

```python
sns.barplot(x='source', y='count', data=orders_source)
```

```python
driver_score_counts = pd.DataFrame(taxi.driver_score.value_counts()).reset_index()
```

```python
driver_score_counts['driver_score'] = (driver_score_counts.driver_score / driver_score_counts.driver_score.sum()) \
                                        .mul(100).round(2)
```

```python
driver_score_counts = driver_score_counts.rename(columns={'index': 'driver_score', 'driver_score': 'percentage'}) \
                    .sort_values('driver_score')
```

```python
driver_score_counts
```

```python
ax = sns.barplot(x='driver_score', y='percentage', data=driver_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Driver score', ylabel='Percentage')
sns.despine()  # ÑƒÐ±Ñ€Ð°Ñ‚ÑŒ Ñ‡Ð°ÑÑ‚ÑŒ Ñ€Ð°Ð¼ÐºÐ¸ Ð³Ñ€Ð°Ñ„Ð¸ÐºÐ°
```

```python
rider_score_counts = pd.DataFrame(taxi.rider_score.value_counts()).reset_index()
```

```python
rider_score_counts['rider_score'] = (rider_score_counts.rider_score / rider_score_counts.rider_score.sum()) \
                                        .mul(100).round(2)
```

```python
rider_score_counts = rider_score_counts.rename(columns={'index': 'rider_score', 'rider_score': 'percentage'}) \
                    .sort_values('rider_score')
```

```python
ax = sns.barplot(x='rider_score', y='percentage', data=rider_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Rider score', ylabel='Percentage')
sns.despine()
```
